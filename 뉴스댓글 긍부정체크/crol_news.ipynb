```python
import csv
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 크롬 드라이버 서비스 생성
driver_service = Service(r'C:\Users\advan\Downloads\chromedriver.exe')

# Chrome 브라우저를 사용하여 WebDriver 객체 생성
driver = webdriver.Chrome(service=driver_service)

# 웹 페이지로 이동
url = 'https://n.news.naver.com/mnews/hotissue/article/comment/047/0002406025?cid=1088007'
driver.get(url)

# 모든 댓글 텍스트를 담을 리스트 초기화
all_comments = []

# 첫 번째 페이지에서 댓글 가져오기
while True:
    # 댓글을 가져오려는 XPath
    xpath = '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li/div[1]/div/div[2]/span[1]'

    # WebDriverWait를 사용하여 댓글이 나타날 때까지 기다림
    wait = WebDriverWait(driver, 10)
    comment_elements = wait.until(EC.presence_of_all_elements_located((By.XPATH, xpath)))

    # 각 댓글 요소에서 텍스트 추출
    comments = [comment.text.strip() for comment in comment_elements]

    # 현재 페이지의 댓글을 전체 댓글 리스트에 추가
    all_comments.extend(comments)

    # 다음 페이지로 이동 (다음 페이지가 없으면 반복 종료)
    try:
        next_page_button = driver.find_element(By.XPATH, '//*[@id="cbox_module"]/div[2]/div[9]/a/span')
        if 'disabled' in next_page_button.get_attribute('class'):
            break
        else:
            next_page_button.click()
    except:
        break

# 모든 댓글 출력
for i, comment in enumerate(all_comments, start=1):
    print(f"댓글 {i}: {comment}")

# 웹 드라이버 종료
driver.quit()
```

```python
import csv
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 크롬 드라이버 서비스 생성
driver_service = Service(r'C:\Users\advan\Downloads\chromedriver.exe')

# Chrome 브라우저를 사용하여 WebDriver 객체 생성
driver = webdriver.Chrome(service=driver_service)

# 웹 페이지로 이동
url = 'https://n.news.naver.com/mnews/hotissue/article/comment/047/0002406025?cid=1088007'
driver.get(url)

# 모든 댓글 텍스트를 담을 리스트 초기화
all_comments = []

# 첫 번째 페이지에서 댓글 가져오기
while True:
    # 댓글을 가져오려는 XPath
    xpath = '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li/div[1]/div/div[2]/span[1]'

    # WebDriverWait를 사용하여 댓글이 나타날 때까지 기다림
    wait = WebDriverWait(driver, 10)
    comment_elements = wait.until(EC.presence_of_all_elements_located((By.XPATH, xpath)))

    # 각 댓글 요소에서 텍스트 추출
    comments = [comment.text.strip() for comment in comment_elements]

    # 공백으로 저장된 댓글 필터링하여 제거
    filtered_comments = [comment for comment in comments if comment != '']

    # 현재 페이지의 필터링된 댓글을 전체 댓글 리스트에 추가
    all_comments.extend(filtered_comments)

    # 다음 페이지로 이동 (다음 페이지가 없으면 반복 종료)
    try:
        next_page_button = driver.find_element(By.XPATH, '//*[@id="cbox_module"]/div[2]/div[9]/a/span')
        if 'disabled' in next_page_button.get_attribute('class'):
            break
        else:
            next_page_button.click()
    except:
        break

# CSV 파일에 댓글 저장 (라벨링 처리)
with open('comments.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['댓글']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    # 헤더 쓰기
    writer.writeheader()
    
    # 댓글 데이터 쓰기
    for comment in all_comments:
        writer.writerow({'댓글': comment})

# 웹 드라이버 종료
driver.quit()
```
